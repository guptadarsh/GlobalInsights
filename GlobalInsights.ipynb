# How has the number of movies released per year changed over the last 20-30 years?

```python
movies_data = n_df[n_df['type']=='Movie']
movies_data.head()
```

```python
changes = movies_data.groupby('release_year')['title'].count()
changes = changes.reset_index()
changes = changes.sort_values(by='title', ascending=False)[:30]
changes
```

```python
changes = movies_data.groupby('release_year')['title'].count()
changes = changes.reset_index()
changes = changes.sort_values(by='title', ascending=False)[:30]
changes
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data = changes,
             x='release_year',
             y='title')
plt.title('Number of Movies released per year')
plt.savefig('Number_of_Movies_released_per_year.png')
plt.show()
```

```python
TVShow_changes = n_df[n_df['type']=='TV Show']
TVShow_changes.head()
```

```python
tvshows_changes = TVShow_changes.groupby('release_year')['title'].count()
tvshows_changes = tvshows_changes.reset_index()
tvshows_changes = tvshows_changes.sort_values(by='title',
                                              ascending=False)[:30]
tvshows_changes
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data = tvshows_changes,
             x='release_year',
             y='title')
plt.title('Number of TV Show released per year')
plt.savefig('Number_of_TVShow_released_per_year.png')
plt.show()
```

# Comparison of tv shows vs. movies.


______________________________________________________________________________

```python
# Top10 Movies
data = unnest_data[unnest_data['type']=='Movie'].groupby('country')['title'].count()
data = data.reset_index()
Top10_Movies = data.sort_values(by='title', ascending=False).head(10)
Top10_Movies
```

```python
# Top10 TVShows
Top10_TVShows = unnest_data[unnest_data['type']=='TV Show'].groupby('country')
Top10_TVShows = Top10_TVShows.agg({'title':'count'}).reset_index()
Top10_TVShows = Top10_TVShows.sort_values(by='title', ascending = False).head(10)
Top10_TVShows
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.barplot(x='country',
            hue='title',
            data=Top10_TVShows)
plt.xticks(rotation=90)
plt.title('Number of TV Shows in Top 10 Country')
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.barplot(x='country',
            y='title',
            data=Top10_TVShows)
plt.xticks(rotation=90)
plt.title('Number of TV Shows in Top 10 Country')
plt.subplot(1,3,3)
sns.barplot(x='country',
            y='title',
            data=Top10_Movies)
plt.xticks(rotation=90)
plt.title('Number of Movies in Top 10 Country')
plt.suptitle('Comparison of TV Shows Vs. Movies')
plt.savefig('Comparison_of_TVShows_Vs._Movies.png')
plt.show()
```

# Analysis of actors/directors of different types of shows/movies.


```python
n_df['cast'] = n_df['cast'].str.split(',')
unnest_data = n_df.explode('cast')
unnest_data.shape
```

```python
unnest_data['director'] = unnest_data['director'].str.split(',')
Unnest_data = unnest_data.explode('director')
Unnest_data.shape
```

```python
Unnest_data.isna().sum()
```

```python
Unnest_data = Unnest_data.fillna({'director':'Unknown', 'cast':'Unknown',
                                  'country':'Unknown', 'date_added':'Unknown',
                                  'rating':'Unknown', 'duration':'0',
                                  'day':'Unknown', 'month':'Unknown'})
```

```python
Unnest_data.isna().sum()
```

```python
# Top 10 Actors in Movies
actor_movies = Unnest_data[Unnest_data['type']=='Movie'].groupby('cast')
actor_movies = actor_movies.agg({'title':'count'})
actor_movies = actor_movies.sort_values(by='title',
                                        ascending=False).head(11).reset_index()
actor_movies
```

```python
# Top 10 Actors in TV Shows
actor_tvshow = Unnest_data[Unnest_data['type']=='TV Show'].groupby('cast')
actor_tvshow = actor_tvshow.agg({'title':'count'})
actor_tvshow = actor_tvshow.sort_values(by='title',
                                        ascending=False).head(11).reset_index()
actor_tvshow
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.barplot(data=actor_tvshow[actor_tvshow['cast'] != 'Unknown'],
            x='cast', y='title')
plt.xticks(rotation=90)
plt.title('Top 10 Actors appears in most TV Shows')
plt.subplot(1,3,3)
sns.barplot(data=actor_movies[actor_movies['cast'] != 'Unknown'],
            x='cast', y='title')
plt.xticks(rotation=90)
plt.title('Top 10 Actors appears in most Movies')
plt.suptitle('Analysis of Actors of different types of TV Shows/Movies')
plt.savefig('analysis_of_Actors_of_different_types_of_TV_Shows_Movies.png')
plt.show()
```

```python
# Top 10 Directors in TV Shows
director_tvshow = Unnest_data[Unnest_data['type']=='TV Show'].groupby('director')
director_tvshow = director_tvshow.agg({'title':'count'})
director_tvshow = director_tvshow.sort_values(by='title',
                                              ascending=False).head(11).reset_index()
director_tvshow
```

```python
# Top 10 Directors in Movies
director_movies = Unnest_data[Unnest_data['type']=='Movie'].groupby('director')
director_movies = director_movies.agg({'title':'count'})
director_movies = director_movies.sort_values(by='title',
                                              ascending=False).head(11).reset_index()
director_movies
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
sns.barplot(data=director_tvshow[director_tvshow['director'] != 'Unknown'],
            x='director', y='title')
plt.xticks(rotation=90)
plt.title('Top 10 Directors appears in most TV Shows')
plt.subplot(1,3,3)
sns.barplot(data=director_movies[director_movies['director'] != 'Unknown'],
            x='director', y='title')
plt.xticks(rotation=90)
plt.title('Top 10 Directors appears in most Movies')
plt.suptitle('Analysis of Directors of different types of TV Shows/Movies')
plt.savefig('Analysis_of_Directors_of_different_types_of_TVShows_Movies.png')
plt.show()
```

# Recommendations

- Company tends to produce more movies than shows, and there has been a rise in both adult and children's content over the past decade. It is suggested to maintain a balance between adult and kids' content.
- The top five movie-producing countries are the United States, India, the United Kingdom, Canada, and France. Similarly, for TV shows, the leading nations are the United States, the United Kingdom, South Korea, Japan, and Canada.
- Adding new movies in January, December, and October is recommended due to festive occasions, with the highest number in these months and the lowest in February and May. Similarly, for TV shows, December is suggested for new releases.
- Based on analysis, keeping movie durations around 100 minutes is recommended, as it aligns with the majority of movies on platform. For TV shows, limiting them to two seasons is suggested to keep the storytelling concise and engaging.- Creating more content in genres like Drama, International Movies, Documentaries, and Comedy is recommended, considering the increasing demand for these genres over the past 30 years.
- It is advised to involve top directors from different countries in content creation, ensuring quality. For instance, directors like Jay Karas and Marcus Raboy are notable in the USA. Similar recommendations can be made for TV shows based on directorial analysis.
- In India, selecting directors like David Dhawan for comedy and Anurag Kashyap for drama is recommended based on their expertise in specific genres, applicable to both movies and TV shows.
- Casting renowned actors like Samuel L. Jackson, Adam Sandler, James Franco, and Nicolas Cage for US productions is suggested. Similar considerations can be made for TV shows based on the cast analysis.
- In India, casting actors like Anupam Kher, Paresh Rawal, and Akshay Kumar for comedy and Naseeruddin Shah, Rajit Kapoor, and Adil Hussain for independent movies is recommended based on their genre-specific experience in both movies and TV shows.
